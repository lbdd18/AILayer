## Análise Bivariada

```python
# Hypotesis: Customers with monthly contract have higher churn rate
# Contingency table of the columns Churn and Contract
pd.crosstab(df_churn['Churn'], df_churn['Contract'], margins=True, margins_name='Total')
```

```python
pd.crosstab(df_churn['Churn'], df_churn['Contract'], margins=True, margins_name='Total', normalize='index')
```

Considerando o publico total, os clientes de contrato mensal representam 55%. Porém, se considerar apenas os clientes que abandonaram o serviço, 88% tinham contrato mensal. Desta forma, podemos afirmar que há uma correspondencia entre o tipo de contrato e o abondono.

```python
# Avaliar a coorelação entre duas variáveis categoricas (qualitativas)

# Executar um teste de hipótese chamado chi-square
# Num testes de hipótese, duas hipóteses são formuladas: 
# - hipótese nula (H0): as duas variáveis são independentes
# - hipótese complementar (H1): as duas variáveis não são independentes

# O testes serve para confirmar ou rejeitar a hipótese nula (H0)
# Quando a probabilidade de observarmos H0 é inferior a 0.05(p-value) -> rejeitamos H0 (nula) e seguimos com H1 (complementar)
```

```python
# Create a dataframe with the crosstab
df_crosstab_churn_contract = pd.crosstab(df_churn['Churn'], df_churn['Contract'])
```

```python
from scipy.stats import chi2_contingency
```

```python
chi_scores_churn_contract = chi2_contingency(df_crosstab_churn_contract)
```

```python
chi_scores_churn_contract
```

```python
scores_churn_contract = pd.Series(chi_scores_churn_contract[0])
```

```python
pvalues_churn_contract = pd.Series(chi_scores_churn_contract[1])
```

```python
# Show numbers with decimals without scientific notation
pd.set_option('display.float_format', lambda x: '%.15f' % x)
```

```python
# Create a dataframe with the results
df_chi_scores_churn_contract = pd.DataFrame({'Qui2': scores_churn_contract, 'p_value': pvalues_churn_contract})
```

```python
df_chi_scores_churn_contract
```

Conforme teste estatistico (Chi-square), o P-value <= 0.05, desta forma rejeitamos a hipotese nula, ou seja, as variaveis não são independentes. Pelo Qui2 alto, podemos afirmar que há uma forte correlação.

```python
# Avaliar a coorelação entre uma variável categórica e uma variável numérica
# Usar o mesmo teste de hipótese chi-square, mas transformar a variável numérica em categórica
# Hipotese: Cliente com menos de 6 meses de contrato possuem maior taxa de Churn

# Criar uma coluna nova no DataFrame com base numa condição de outra coluna
# Coluna = 'TimeLess6Months'
import numpy as np
```

```python
# Create a new column TimeLess6Months
df_churn['TimeLess6Months'] = np.where(df_churn['tenure'] < 6, 'Yes', 'No') 
```

```python
df_churn.head()
```

```python
df_crosstab_churn_tenure = pd.crosstab(df_churn['Churn'], df_churn['TimeLess6Months'])
```

```python
# Calcular os scores e P-Values
chi_scores_churn_tenure = chi2_contingency(df_crosstab_churn_tenure)
```

```python
scores_churn_tenure = pd.Series(chi_scores_churn_tenure[0])
pvalues_churn_tenure = pd.Series(chi_scores_churn_tenure[1])
```

```python
df_crosstab_churn_tenure = pd.DataFrame({'Qui2': scores_churn_tenure, 'p_value': pvalues_churn_tenure})
```

```python
df_crosstab_churn_tenure
```

Como p-value <= 0.05, rejeitamos a hipótese nula, ou seja, as variáveis não são independentes. Com base no Qui2 resultante dos dois testes estatisticos a coorelação churn/ tempo de contracto <6 meses é menor que a do churn/tipo de contrato.

```python
# Correlação entre variáveis numéricas
# Tenure com total charges
# A intuição é que clientes com mais tempo de contrato devem ter um valor total de cobrança maior

# Correlação de Pearson usando pearson
df_churn['tenure'].corr(df_churn['TotalCharges'], method='spearman')
```

há uma forte coorelação entre tenure e total charges usando metodos estatiticos de coorelação.
