# Import libraries

```python
import pandas as pd
```

# Data collection

## Load data

```python
df_customers = pd.read_csv('datasets/churn_customers.csv')
df_services = pd.read_csv('datasets/churn_services.csv')
df_contracts = pd.read_csv('datasets/churn_contracts.csv')
```

```python
df_customers.tail()
```

```python
df_customers.info()
```

## Data transformation

```python
df_contracts.info()
```

```python
#df_contracts['TotalCharges'] = df_contracts['TotalCharges'].astype(float)

df_contracts['TotalCharges'] = pd.to_numeric(df_contracts['TotalCharges'], errors='coerce')
```

```python
df_contracts.info()
```

```python
df_customers.rename(columns={'SeniorCitizen': 'Above65yo'}, inplace=True)
```

```python
df_customers.info()
```

## Merge dataframes

```python
len(df_customers)
len(df_services)
len(df_contracts)
```

```python
df_temp = df_customers.merge(df_services, on='customerID')
```

```python
df_temp.head()
```

```python
df_contracts.rename(columns={'customerID': 'IDCliente'}, inplace=True)
```

```python
df_contracts.info()
```

```python
df_churn = df_temp.merge(df_contracts, left_on=['customerID'], right_on=['IDCliente'])
```

```python
df_churn.info()
```

## Missing data

```python
df_churn.isna().sum()
```

```python
df_churn['TotalCharges'].isna().sum()
```

```python
df_churn.drop(columns=['TotalCharges'], axis=1)
```

```python
# Drop columns with NaN values
df_churn.dropna(axis=1)
```

```python
# Drop rows with NaN values
df_churn.dropna(axis=0)
```

```python
df_churn.info()
```

```python
df_churn.fillna(0)
```

```python
df_churn.fillna(value={'TotalCharges': df_churn['TotalCharges'].median()}, inplace=True)  # Fill NaN values with the mean
```

```python
df_churn.info()
```

```python
df_churn.fillna(value={'gender': df_churn['gender'].mode()}, inplace=True)  # Fill NaN values with the mode
```
